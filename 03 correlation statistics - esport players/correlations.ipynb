```python
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
# load file
data = pd.read_csv("SDM_player_gdp_pop_int.csv")
# data type settings
data = pd.DataFrame(data)
data["GPI_CNT_PLAYER_ID"] = data["GPI_CNT_PLAYER_ID"].astype(int)
data["GPI_CNT_PLAYERS_ON_1MIL_POP"] = data["GPI_CNT_PLAYERS_ON_1MIL_POP"].astype(float)
data["GPI_AVG_GDP_PER_CAPITA"] = data["GPI_AVG_GDP_PER_CAPITA"].astype(float)
data["GPI_AVG_POPULATION"] = data["GPI_AVG_POPULATION"].astype(float)
data["GPI_AVG_POPULATION_AGED_15TO29"] = data["GPI_AVG_POPULATION_AGED_15TO29"].astype(float)
data["GPI_AVG_SHARE_OF_INDIVIDUALS_USING_INTERNET"] = data["GPI_AVG_SHARE_OF_INDIVIDUALS_USING_INTERNET"].astype(float)
# text columns must be excluded
data.drop(["GPI_COUNTRY","GPI_CONTINENT"], axis='columns', inplace=True)

# Rename columns
data = data.rename(columns={"GPI_CNT_PLAYER_ID": "No. of players", "GPI_CNT_PLAYERS_ON_1MIL_POP": "No. of players/mil. inhabitants", "GPI_AVG_GDP_PER_CAPITA": "Avg. GDP per capita", "GPI_AVG_POPULATION": "Avg. population", "GPI_AVG_POPULATION_AGED_15TO29": "Avg. population aged 15-29 years", "GPI_AVG_SHARE_OF_INDIVIDUALS_USING_INTERNET": "Avg. perc. of inhabitants using the Internet"})
# print the first records
print(data.head())
```

```python
sns.scatterplot(data=data, x="Avg. population", y="No. of players", color="blue")
```

```python
sns.scatterplot(data=data, x="Avg. population aged 15-29 years", y="No. of players", color="blue")
```

```python
sns.scatterplot(data=data, x="Avg. GDP per capita", y="No. of players/mil. inhabitants", color="blue")
```

```python
sns.scatterplot(data=data, x="Avg. perc. of inhabitants using the Internet", y="No. of players/mil. inhabitants", color="blue")
```

```python
# Normal distribution test:
# Normal distribution is always tested for one column of the table.
# If the p-value is less than a certain pre-set threshold (0.05), then the data do not come from a normal distribution.

print(stats.normaltest(data["No. of players"]))
print(stats.normaltest(data["No. of players/mil. inhabitants"]))
print(stats.normaltest(data["Avg. GDP per capita"]))
print(stats.normaltest(data["Avg. population"]))
print(stats.normaltest(data["Avg. population aged 15-29 years"]))
print(stats.normaltest(data["Avg. perc. of inhabitants using the Internet"]))
```

```python
## FINDING THE CORRELATION VALUE
# statistic - the higher it is the stronger is the correlation
print(stats.kendalltau(data["Avg. population"], data["No. of players"]))
print(stats.kendalltau(data["Avg. perc. of inhabitants using the Internet"], data["No. of players/mil. inhabitants"]))
```

```python
## FINDING THE CORRELATION VALUE
# statistic - the higher it is, the stronger is the correlation
print(stats.kendalltau(data["Avg. GDP per capita"], data["No. of players/mil. inhabitants"]))
print(stats.kendalltau(data["Avg. population aged 15-29 years"], data["No. of players"]))
```
