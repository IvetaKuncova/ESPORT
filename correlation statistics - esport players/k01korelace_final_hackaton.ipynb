```python
data = pd.read_csv("SDM_player_gdp_pop_int.csv")

##"GPI_COUNTRY","GPI_CONTINENT","GPI_CNT_PLAYER_ID","GPI_AVG_GDP_PER_CAPITA","GPI_AVG_POPULATION","GPI_AVG_POPULATION_AGED_15TO29","GPI_CNT_PLAYERS_ON_1MIL_POP","GPI_AVG_SHARE_OF_INDIVIDUALS_USING_INTERNET"

data = pd.DataFrame(data)
data["GPI_CNT_PLAYER_ID"] = data["GPI_CNT_PLAYER_ID"].astype(int)
data["GPI_CNT_PLAYERS_ON_1MIL_POP"] = data["GPI_CNT_PLAYERS_ON_1MIL_POP"].astype(float)
data["GPI_AVG_GDP_PER_CAPITA"] = data["GPI_AVG_GDP_PER_CAPITA"].astype(float)
data["GPI_AVG_POPULATION"] = data["GPI_AVG_POPULATION"].astype(float)
data["GPI_AVG_POPULATION_AGED_15TO29"] = data["GPI_AVG_POPULATION_AGED_15TO29"].astype(float)
data["GPI_AVG_SHARE_OF_INDIVIDUALS_USING_INTERNET"] = data["GPI_AVG_SHARE_OF_INDIVIDUALS_USING_INTERNET"].astype(float)

data.drop(["GPI_COUNTRY","GPI_CONTINENT"], axis='columns', inplace=True)

# Přejmenování dvou sloupců najednou
data = data.rename(columns={"GPI_CNT_PLAYER_ID": "Počet hráčů", "GPI_CNT_PLAYERS_ON_1MIL_POP": "Počet hráčů/mil. obyvatel", "GPI_AVG_GDP_PER_CAPITA": "Prům. výše HDP na obyvatele", "GPI_AVG_POPULATION": "Prům. počet obyvatel", "GPI_AVG_POPULATION_AGED_15TO29": "Prům. počet obyv. ve věku 15-19 let", "GPI_AVG_SHARE_OF_INDIVIDUALS_USING_INTERNET": "Prům. procento obyvatel užívajících internet"})


print(data.head())
```

```python
# test na normální rozdělení dat:
# normální rozdělení se vždy testuje pro jeden sloupec tabulky
# sleduji hodnotu p-value, pokud je menší než 0,05, pak 

"""D’Agostino a Pearson test
Na základě kombinace zkosení a špičatosti křivky dat testuje to, zda vzorek pochází z normálního rozdělení.  
Pokud je p-hodnota menší než určitý předem stanovený práh (často 0,05), pak data nepochází z normálního rozdělení. 
statisticfloat or array
s^2 + k^2, where s is the z-score returned by skewtest and k is the z-score returned by kurtosistest.
pvaluefloat or array
A 2-sided chi squared probability for the hypothesis test.
P-VALUE!!!
 p-hodnota znamená pravděpodobnost, že data, na kterých je test prováděn, pocházejí z normálního rozložení. 
Pokud je p-hodnota menší než zvolená hladina významnosti 0,05, 
pak je pravděpodobné, že data pocházejí z jiného rozložení, než je normální rozložení.

statistic: hodnota testového statistického ukazatele
Testový statistický ukazatel je hodnota, která se využívá k porovnání dat se standardní normální distribucí. Pokud je hodnota testového statistického ukazatele dostatečně velká, pak je pravděpodobné, že data pocházejí z jiné populace, než je normální populace.

P-hodnota pak vyjadřuje pravděpodobnost, že by se hodnota testového statistického ukazatele, která je alespoň tak velká jako pozorovaná, objevila náhodou, pokud by data pocházela z normální populace.

Obecně platí, že pokud je p-hodnota menší než zvolená hladina významnosti, pak se považuje za statisticky významnou. To znamená, že je nepravděpodobné, že by se pozorovaná data objevila náhodou, pokud by data pocházela z normální populace.

Výběr hladiny významnosti je subjektivní rozhodnutí. Obecně platí, že hladina významnosti 0,05 se považuje za standardní. To znamená, že pokud je p-hodnota menší než 0,05, pak se považuje za statisticky významnou.
"""
print(stats.normaltest(data["Počet hráčů"]))
print(stats.normaltest(data["Počet hráčů/mil. obyvatel"]))
print(stats.normaltest(data["Prům. výše HDP na obyvatele"]))
print(stats.normaltest(data["Prům. počet obyvatel"]))
print(stats.normaltest(data["Prům. počet obyv. ve věku 15-19 let"]))
print(stats.normaltest(data["Prům. procento obyvatel užívajících internet"]))

#NormaltestResult(statistic=225.09172698161382, pvalue=1.3241967721243712e-49) 0.000000000000000000000000000013241967721243712
#NormaltestResult(statistic=107.58764136839392, pvalue=4.3415071076559595e-24) 0.0000000000000000043415071076559595
#NormaltestResult(statistic=38.56032073757092, pvalue=4.233823426609993e-09) 0.00000000004233823426609993
#NormaltestResult(statistic=202.5601007432329, pvalue=1.0342677828014017e-44) 0.000000000000000000000000000000010342677828014017
#NormaltestResult(statistic=203.0852987523566, pvalue=7.954038670598638e-45) 0.00000000000000000000000000000007954038670598638
#NormaltestResult(statistic=9.833171528727362, pvalue=0.0073240943909007915)
```

```python
## ZJIŠTENÍ KORELACE - Počet hráčů (statistics viz data.corr)
print(stats.kendalltau(data["Prům. výše HDP na obyvatele"], data["Počet hráčů"]))
print(stats.kendalltau(data["Prům. počet obyvatel"], data["Počet hráčů"]))
print(stats.kendalltau(data["Prům. počet obyv. ve věku 15-19 let"], data["Počet hráčů"]))
print(stats.kendalltau(data["Prům. procento obyvatel užívajících internet"], data["Počet hráčů"]))

#statistic - čím je vyšší tím je korelace silnější
# pvalue když je pod 0,05 je statisticky významná, pravděpodobost, že tam korelace opravdu je  a není to náhoda

# SignificanceResult(statistic=0.3113288291512198, pvalue=2.996594231966616e-07) 0.0000002996594231966616
# SignificanceResult(statistic=0.3702218659989922, pvalue=1.1089236960257302e-09) 0.000000000011089236960257302
# SignificanceResult(statistic=0.30977218500546366, pvalue=3.4318329221465256e-07) 0.00000034318329221465256
# SignificanceResult(statistic=0.3800806122554475, pvalue=3.971690688428846e-10) 0.0000000000003971690688428846
```
